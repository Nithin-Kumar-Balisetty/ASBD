Question 2

```python
def firstDigit(n) :
    while n >= 10: 
        n = n / 10;
    return int(n)

dataset=[22, 21, 24, 19, 27, 28, 24, 25, 29, 28, 26, 31, 28, 27, 22, 39, 20, 10, 26, 24, 27, 28, 26, 28, 18, 32, 29, 25, 31, 27]
first_digits = [ firstDigit(i) for i in dataset]
y = pd.Series(dataset)
fig, ax = stemgraphic.stem_graphic(y)
q1= np.quantile(dataset,.25)
q3 = np.quantile(dataset, .75)
lower_outlier = q1 - 1.5*(q3-q1)
upper_outlier = q3 + 1.5*(q3-q1)
print("Outliers are :")
outliers =[i for i in dataset if (i>=upper_outlier) or (i<=lower_outlier) ]
print(outliers)
```

```python
col=[]
#plt.ylabel('Data')   # for label at y-axis
#plt.xlabel('stems')   # for label at x-axis
#plt.xlim(0,5)   # limit of the values at x axis
#plt.stem(first_digits, dataset)   # required plot

for i in dataset:
    if i in outliers:
        col.append("red")
    else:
        col.append('black')
for i in range(len(dataset)):
    plt.scatter(first_digits[i], dataset[i], c = col[i],linewidth = 0)
print("Red Points indicate outliers")      
plt.show()

```

```python
water=[3.2, 3.5, 3.6, 2.5, 2.8, 5.9, 2.9, 3.9, 4.9, 6.9, 7.9, 8.0, 3.3, 6.6, 4.4]
bevarage = [3.2, 3.5, 3.6, 2.5, 2.8, 5.9, 2.9, 3.9, 4.9, 6.9, 7.9, 8.0, 3.3, 6.6, 4.4]

main_series = pd.Series({"Water Consumption" : water,"Beverage Consumption" : bevarage})

seaborn.set_theme()
seaborn.kdeplot(data=main_series,x="Water Consumption")
seaborn.rugplot(data=main_series, x="Water Consumption",height=0.05)
```

```python
seaborn.set_theme()
seaborn.kdeplot(data=main_series,x="Beverage Consumption")
seaborn.rugplot(data=main_series, x="Beverage Consumption",height=0.05)
```

```python
print("Mean Water Consumption : "+str(statistics.mean(water)))
print("Mean Beverage Consumption : "+str(statistics.mean(bevarage)),end="\n\n")

print("Median Water Consumption : "+str(statistics.median(water)))
print("Median Beverage Consumption : "+str(statistics.median(bevarage)),end="\n\n")

print("Mode Water Consumption : "+str(statistics.mode(water)))
print("Mode Beverage Consumption : "+str(statistics.mode(bevarage)),end="\n\n")

print("Skewness Water Consumption : "+str(skew(water)))
print("Skewness Beverage Consumption : "+str(skew(bevarage)))
```

```python
fuel =[3.6,6.7,9.8,11.2,14.7]
mass= [0.45,0.91,1.36,1.81,2.27]

plt.scatter(mass,fuel)
plt.xlabel("Mass of the Car")
plt.ylabel("Fuel used")
plt.plot(np.unique(mass), np.poly1d(np.polyfit(mass, fuel, 1))(np.unique(mass)))
plt.show()
c, _ = pearsonr(mass, fuel)
print("Pearson Correlation : "+str(c))
if c>0:
    print("Positve Correlation")
elif c==0:
    print("Nuetral Correlation")
else: 
    print("Negative Correlation")
```
